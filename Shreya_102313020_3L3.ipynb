# Step - 1

```python
import numpy as np
rno = 102313020
data = df['no2'].dropna()
x = data.values
r=rno
ar = 0.05 * (r % 7)
br = 0.3 * (r % 5 + 1)
z = x + ar * np.sin(br * x)

print(f"Step 1 Complete.")
print(f"Parameters used: ar = {ar}, br = {br}")
```

# Step - 2

```python
mean = np.mean(z)
sigma = np.std(z)
lambda_val = 1 / (2 * sigma**2)
c_val = 1 / (sigma * np.sqrt(2 * np.pi))

print(f"rno={rno}")
print(f"mean={mean}")
print(f"lambda_val={lambda_val}")
print(f"c_val={c_val}")
```
